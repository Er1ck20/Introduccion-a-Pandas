```python
# Importamos
import pandas as pd
```

```python
# Leemos el DataFrame
df = pd.read_csv('StudentsPerformance.csv')

# Mostramos el dataframe
df
```

# Crear Index

```python
# Importamos
import numpy as np
import random
```

```python
# Creamos valores no repetitivos para los index
# arange - Nos devuelve un rango de los valores asignados.
n_index = np.arange(0, 1000)
n_index
```

```python
# Desordenar los index
#shuffle () - Nos baraja o desordena los datos
random.shuffle(n_index)
```

```python
# Creamos una nueva columna con los nuevos index
df['new index'] = n_index
```

```python
df
```

# Set_index()

```python
# fijar columna new_index como index, ponerla al principio.
# Inplace = True - Guardamos/actualizamos los datos
df.set_index('new index', inplace=True)
```

```python
df
```

# Sort_index()

```python
# Ordenar DataFrame por new index
df.sort_index()
```

```python
# Ordenar de manera descendiente por index
df.sort_index(ascending=False)
```
