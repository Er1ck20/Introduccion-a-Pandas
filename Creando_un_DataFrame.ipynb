# Creando un DataFrame

```python
# Importamos
import numpy as np
import pandas as pd
```

## Crear DataFrame con Array

```python
# Primera Forma
# Creando el array
data = np.array([[2,3], [2,5], [3,2]])
```

```python
# Creando un dataframe - Usamos el Metodo DataFrame, agregamos el array ,el argumento index y el columns, podemos cambiarle el nombre a dichos argumentos para mostrarlos en el dataframe
df = pd.DataFrame(data, index=['Fila 1', 'Fila 2', 'Fila 3'], columns=['Columna 1', 'Columna 2'])
```

```python
# Mostrando el dataframe
df
```

## Crear DataFrame con forma de Listas (Arrays)

```python
# Segunda forma
# Creando un DataFrame en forma de lista
data2 = [[1, 2],  [3, 1]]
```

```python
# Creando un dataframe
df = pd.DataFrame(data2, index=['Fila 1', 'Fila 2'], 
                  columns=['Col 1', 'Col 2'])
```

```python
# Mostrando el DataFrame
df
```

# Crear DataFrame Usando Diccionarios

```python
# Listas que usaremos
Jugadores = ['Lebron James', 'Stephen Curry', 'James Harden', 'Russell Westbrook']
Sueldo = ['48.1 millones', '47.4 millones', '47.1 millones', '44.5 millones']
```

```python
# Almacenando la lista en un diccionario
dict_jugadores = {'Jugadores': Jugadores, 'Sueldo': Sueldo}
```

```python
# Creando el DataFrame
df = pd.DataFrame(dict_jugadores)
```

```python
# Mostrando el DataFrame
df
```

```python
# Leer el archivo CSV - Usamos el metodo read_csv()
# Los datos almacenado dentro del csv se le conoce como dataset
df = pd.read_csv('StudentsPerformance.csv')
```
