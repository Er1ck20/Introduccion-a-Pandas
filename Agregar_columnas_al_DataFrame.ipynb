# Agregar una nueva columna al DataFrame usando Assign()

### Cuando usar assign()?
#### Es mas conveniente usarlo cuando queremos agregar multiples columnas en una sola linea de codigo.
#### Tambien cuando se necesita sobreescribir los valores de una columna existente(BUENA PRACTICA).
##### Devuelve un nuevo  objeto (copia) con todas las columnas originales ademas de las nuevas.

```python
# Importamos
import numpy as np
import pandas as pd
```

```python
# Creamos numeros aleatotorios para nuestras nuevas columnas
numero1 = np.random.randint(70, 100, size=1000)
numero2 = np.random.randint(80, 100, size=1000)
```

```python
# Creamos unas series usando los numeros aleatorios
# Series - Es una lista(array) de una dimension.
# Arange - Es un rango del 0 al 1000.
series1 = pd.Series(numero1, index=np.arange(0, 1000))
series2 = pd.Series(numero2, index=np.arange(0, 1000))
```

```python
series2
```

```python
# Usando assign() para agregar multiples columnas
# assign() devuelve es una copia del objeto, si la queremos guardar
# debemos sobreescribirla, ejemplo df = df.assign()
df = df.assign(numero1 = series1, numero2 = series2) # Esta sobreescrita
```

```python
df.head(10)
```

# Agregando una columna al DataFrame con el metodo Insert()
