```python
root = 'https://www.football-data.co.uk/mmz4281/'
```

```python
# Creamos una lista de ligas
lista_ligas = ['SC0', 'SC1', 'SC2']
# Lista vacia
frames = []

# Bucle for en la lista de ligas
for ligas in lista_ligas:
    # Bucle for - Realizaremos un bucle a un range de temporadas.
    for season in range(19, 23):
        
        # OJO: Estamos convirtiendo las temporadas(season) en una cadena de texto y estamos sumando la misma variable +1, asi podemos agregar las temporadas juntas(eje: 2021)
        df = pd.read_csv( root + str(season) + str(season+1)  + '/' + ligas + '.csv')

        # Creamos una nueva columna, para identificar a que temporada pertenece cada df.
        df.insert(1, 'Season', season)

        # Agregaremos el dataframe con el metodo append a la lista vacia frames.
        frames.append(df)
```

## 3. Organizar la data en un dict(Diccionario)
* Para organizar todos los csv que hemos extraido de la web, usaremos los diccionarios que nos ayudaran a identificar facilmente que archivo/elemento queremos ver/acceder sin la necesidad de hacer una indexación de listas.
* OJO: Nos evita hacer esto -> frames[1]

```python
# Crearemos un diccionario con el nombre original de la liga como llave
dict_paises = {
    'Scotland Premier League': 'SC0',
    'Scoltland League 1': 'SC1',
    'Scoltland League 2': 'SC2'
}
```

```python
# Obtener un elemento del diccionario
dict_paises['Scotland Premier League']
```

```python
# Diccionario vacio
dict_historia_data = {}

# Hacer un bucle sobre las llaves del diccionario
for ligas in dict_paises:
    # Lista vacia
    frames = []
    # Bucle for - Realizaremos un bucle y insertaremos un range de temporadas.
    for season in range(19, 23):        
        df = pd.read_csv( root + str(season) + str(season+1)  + '/' + dict_paises[ligas] + '.csv')
        df.insert(1, 'Season', season)
        frames.append(df)
    
    # Metodo pd.concat() - Nos ayuda a concatenar elementos, en este caso todos los df
    df_concat = pd.concat(frames)
    
    # Agregamos las ligas al diccionario vacio.
    dict_historia_data[ligas] = df_concat
```
