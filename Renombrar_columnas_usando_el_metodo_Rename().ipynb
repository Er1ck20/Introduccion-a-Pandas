```python
# Importamos
import pandas as pd
```

```python
# Leemos el dataframe
df = pd.read_csv('StudentsPerformance.csv')
# Mostramos el dataframe
df
```

# Renombrar columnas con el metodo rename()

```python
# Renombrar una columna y sobreescribir los valores del DataFrame
# Para renombrar la columna usamos columns y usamos un diccionario asignando el nombre a cambiar como la llave y el nuevo nombre como el valor
df.rename(columns={'gender':'Gender'})
```

```python
# Renombrar una columna y actualizar el dataframe usando inplace
df.rename(columns={'reading score':'Calificacion de lectura'}, inplace=True)
```

```python
df
```

```python
# Renombrar varias columnas y actualizar el dataframe
df.rename(columns={'math score': 'Calificacion de matematica', 'writing score': 
                   'Calificacion de escritura'}, inplace = True)
```

```python
df
```

# Renombrar index

```python
# Renombrar index 0, 1 y 2 y actualizar el dataframe con inplace
# Para renombrar el indice usamos el argumento index 
df.rename(index={0: '000', 1:'001', 2:'002'}, inplace=True)
```
